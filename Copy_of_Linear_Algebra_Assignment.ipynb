## 1.1 Create a two-dimensional vector and plot it on a graph

```
import matplotlib.pyplot as plt
import numpy as np
```

```
a = np.array([.1, .1])
b = np.array([.4, .6])
c = a / b


plt.arrow(0,0, a[0],a[1], head_width=.02, head_length=0.01, color='r')
plt.arrow(0,0, b[0], b[1], head_width=.02, head_length=0.01, color='orange')
plt.arrow(0,0, c[0], c[1], head_width=.02, head_length=0.01, color='green')

plt.show()
```
